```python
import requests
api_url = 'https://api.bigw.com.au/api/stores/v0/list' 
response = requests.get(api_url)


if response.status_code == 200:
    data = response.json()

else:
    print(f"Request failed with status code: {response.status_code}")
```

```python
import pandas as pd

fd=pd.DataFrame(data)
```

```python
fd
df=fd.T
df
```

```python
store_info=df
store_info=store_info.drop(columns=['enabledForInstore','enabledForPickup','serviceDesk','driveUp','openingHours','tradingHours','services'])
store_info
```

```python
postcode=store_info['address'].apply(lambda x: x['postcode'])
suburb=store_info['address'].apply(lambda x: x['suburb'])
street=store_info['address'].apply(lambda x: x['street'])
state=store_info['address'].apply(lambda x: x['state'])


latitude=store_info['location'].apply(lambda x: x['lat'])
longitude=store_info['location'].apply(lambda x: x['lng'])
```

```python
store_info['street'] = street
store_info['suburb'] = suburb
store_info['state'] = state
store_info['postcode'] = postcode
store_info['latitude'] = latitude
store_info['longitude'] = longitude
store_info=store_info.drop(columns=['address','location'])
store_info.rename(columns = {'name':'storeName'}, inplace = True)
store_info.rename(columns = {'id':'brandName'}, inplace = True)
```

```python
store_info["brandName"]="BIG W"
store_info['country']='Australia'
index = pd.Index(range(1, 178, 1))
store_info = store_info.set_index(index)
store_info = store_info.reindex(['brandName','storeName','phoneNumber','street','suburb','state','country','postcode','latitude','longitude'], axis=1)
store_info['phoneNumber'] = store_info['phoneNumber'].str.replace(" ", "")
```

```python
url2 = "https://www.choicethediscountstore.com.au/store-finder/"
response2 = requests.get(url2)
response2.text
soup= BeautifulSoup(response2.text, 'html.parser')
soup
```

```python
data=soup.find_all('div',class_='col-12 col-md-4')
c=0
store_name2 = []
phone2= []
suburb2 = []
street2=[]
post2 = []
country2 = []
timings2 =[]
state2=[]


for d in data:
  my_list = d.text.split('\n')
  my_list = list(filter(None,my_list))
  
  if(c%2==0):
    store_name2.append("Choice The Discount Store - " + my_list[0])
    if(len(my_list) == 2):

      phone2.append(my_list[1].replace(" ", ""))
    else:
      phone2.append("")
  else:
    if(c==1):
      street2.append(my_list[0].split(', ')[1])
    else:
      street2.append(my_list[0])
    suburb2.append(my_list[1].split(',')[0])
    if(len(my_list) == 3):
      state2.append((my_list[1].split(',')[1]).split(" ")[1])
      post2.append((my_list[1].split(',')[1]).split(" ")[2])
      country2.append(my_list[1].split(', ')[2])
      timings2.append((my_list[2].split('C')[0]).replace("PM", "PM "))
    else:
      state2.append('N/A')
      post2.append('N/A')
      country2.append('Australia')
      timings2.append('N/A')
  c = c+1
suburb2[26]='Windsor'
street2[26]='N/A'
phone2[26]='N/A'
print((phone2))
print((store_name2))
print((street2))
print(len(suburb2))
print((post2))
print((state2))
print((country2))
print((timings2))
print(c)
```
